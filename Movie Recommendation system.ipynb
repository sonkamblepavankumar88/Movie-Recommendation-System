### import Abstract Syntax Tree

```python
import ast
```

#### An Abstract Syntax Tree, or AST, is a tree representation of the source code of a computer program that conveys the structure of the source code. Each node in the tree represents a construct occurring in the source code.

### import warnings to ignore all the warnings 

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

### Converting stringified lists in 'genres', 'keywords', 'cast', and 'crew' columns into actual lists using the ast.literal_eval() function.

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
        return L
```

### This function is specifically designed to extract a maximum of 3 'name' values from a list of dictionaries.

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
    return l
```

### This function is typically used to extract the names of directors from a column that contains a stringified list of dictionaries, where each dictionary represents a person's job and name associated with a particular movie.

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

###  this code splits the 'overview' column of the 'movies' dataframe, treating each string value as a sentence or collection of words, and converts it into a list of individual words. This can be useful for further processing or analysis that requires working with individual words rather than entire sentences.

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace("", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace("", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace("", "") for i in x] if x is not None else [])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace("", "") for i in x] if x is not None else [])
```

### these lines of code remove any occurrence of empty strings ("") within the lists present in the 'genres', 'keywords', 'cast', and 'crew' columns of the 'movies' dataframe.

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

### this code combines the information from different columns into a single column named 'tags', which can be used as a consolidated representation of various aspects of a movie, including its overview, genres, keywords, cast, and crew.

```python
movies.head()
```

```python
new = movies[['movie_id', 'title', 'tags']]
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: ' '.join(x))
```

### this code converts the 'tags' column from a list of words into a single string by joining the individual words with a space between them

```python
new['tags']=new['tags'].apply(lambda X:X.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
import nltk
```

### Natural Language Toolkit, is a Python package that you can use for NLP. A lot of the data that you could be analyzing is unstructured data and contains human-readable text. Before you can analyze that data programmatically, you first need to preprocess it.

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

### this code sets up the PorterStemmer object 'ps', which can be used to perform stemming operations on words in text data.

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

### this code defines a function that applies stemming to the words in a given text. The function uses the PorterStemmer object 'ps' to perform the stemming operation and returns the stemmed text as output.

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

### the cosine_similarity function is used to calculate the cosine similarity between vectors or matrices, providing a measure of similarity between them.

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])),reverse = True, key = lambda x:x[1])[1:6]
```

###  the code sorts the similarity values of the first row in descending order and selects the top 5 elements (excluding the first element). These elements correspond to the indices of the most similar movies to the first movie in the similarity matrix.

```python
def recommend(movie):
    movie_index = new[new['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances )),reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new.iloc[i[0]].title)
```

### the 'recommend' function retrieves the index of a given movie, calculates the similarity values between that movie and all other movies, selects the top 5 similar movies, and prints their titles as recommendations.

### Check the codes are working correctly

```python
recommend('Avatar')
```
